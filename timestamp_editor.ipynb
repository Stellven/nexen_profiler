# Timestamp Editor (Per-File)

Edit modification times for individual files in `data/read/` and `data/write/`.

Workflow:
1. Run the setup cell to load files.
2. Edit the `overrides` dict (one file at a time) or call `set_override(...)`.
3. Run the apply cell to set mtimes.
4. Run the verify cell to spot-check results.


```python
from __future__ import annotations

import os
from datetime import datetime, timezone
from pathlib import Path

BASE_DIR = Path('.')
DATA_DIR = BASE_DIR / 'data'
READ_DIR = DATA_DIR / 'read'
WRITE_DIR = DATA_DIR / 'write'

print('Read dir:', READ_DIR.resolve())
print('Write dir:', WRITE_DIR.resolve())

def list_files(root: Path):
    if not root.exists():
        return []
    return [p for p in root.rglob('*') if p.is_file()]

read_files = list_files(READ_DIR)
write_files = list_files(WRITE_DIR)

print('Read files:', len(read_files))
print('Write files:', len(write_files))
```

```python
def parse_dt(value: str) -> datetime:
    formats = [
        '%Y-%m-%d %H:%M:%S',
        '%Y-%m-%d %H:%M',
        '%Y-%m-%d',
        '%Y-%m-%dT%H:%M:%S',
        '%Y-%m-%dT%H:%M',
    ]
    for fmt in formats:
        try:
            return datetime.strptime(value, fmt).replace(tzinfo=timezone.utc)
        except ValueError:
            continue
    raise ValueError(f'Unsupported datetime format: {value}')

def set_timestamp(path: Path, dt: datetime) -> None:
    ts = dt.timestamp()
    os.utime(path, (ts, ts))

def current_ts(path: Path) -> str:
    dt = datetime.fromtimestamp(path.stat().st_mtime, tz=timezone.utc)
    return dt.strftime('%Y-%m-%d %H:%M:%S')

def build_overrides(paths: list[Path]) -> dict[str, str]:
    return {str(p): current_ts(p) for p in paths}

def set_override(overrides: dict[str, str], stem: str, date_str: str, paths: list[Path] | None = None) -> None:
    if paths is None:
        paths = read_files + write_files
    matches = [p for p in paths if p.stem == stem]
    if not matches:
        raise ValueError(f'No file found with stem: {stem}')
    if len(matches) > 1:
        names = ', '.join(p.name for p in matches)
        raise ValueError(f'Ambiguous stem {stem}. Matches: {names}')
    dt = parse_dt(date_str)
    overrides[str(matches[0])] = dt.strftime('%Y-%m-%d %H:%M:%S')

def print_overrides(overrides: dict[str, str]) -> None:
    for i, (k, v) in enumerate(overrides.items()):
        print(f"{i}. {k}: {v}")

def set_override_index(overrides: dict[str, str], index: int, date_str: str) -> None:
    keys = list(overrides.keys())
    if index < 0 or index >= len(keys):
        raise IndexError(f'Index out of range: {index}')
    if date_str == '':
        overrides[keys[index]] = ''
        return
    dt = parse_dt(date_str)
    overrides[keys[index]] = dt.strftime('%Y-%m-%d %H:%M:%S')
```

```python
# Per-file overrides: edit values below.
# You can delete keys or set a value to '' to skip that file.

# Helper: set_override_index(overrides, 0, 'YYYY-MM-DD HH:MM') (use '' to skip)

# Helper: set_override(overrides, 'file_name_without_ext', 'YYYY-MM-DD HH:MM')

overrides = build_overrides(read_files + write_files)
print_overrides(overrides)

# read
set_override_index(overrides, 0, '2026-03-02')
set_override_index(overrides, 2, '2026-03-01')
set_override_index(overrides, 3, '2026-02-15')
set_override_index(overrides, 4, '2026-02-15')
set_override_index(overrides, 5, '2026-02-12')
set_override_index(overrides, 7, '2026-02-10')
set_override_index(overrides, 8, '2026-03-04')
set_override_index(overrides, 9, '2026-01-30')
set_override_index(overrides, 10, '2026-02-26')
set_override_index(overrides, 11, '2026-03-01')
set_override_index(overrides, 12, '2025-12-31')
set_override_index(overrides, 13, '2026-02-20')
set_override_index(overrides, 14, '2026-03-04')
set_override_index(overrides, 15, '2026-02-28')
set_override_index(overrides, 17, '2026-02-06')
# write
set_override_index(overrides, 18, '2026-02-15')
set_override_index(overrides, 19, '2026-02-20')
set_override_index(overrides, 20, '2026-02-25')
set_override_index(overrides, 21, '2026-02-28')
set_override_index(overrides, 22, '2026-03-02')
set_override_index(overrides, 23, '2026-02-04')

```

```python
# Apply per-file overrides

changed = 0
for path_str, ts_str in overrides.items():
    if not ts_str:
        continue
    path = Path(path_str)
    if not path.exists():
        print('Missing:', path)
        continue
    new_dt = parse_dt(ts_str)
    old_dt = datetime.fromtimestamp(path.stat().st_mtime, tz=timezone.utc)
    if abs((new_dt - old_dt).total_seconds()) < 1:
        continue
    set_timestamp(path, new_dt)
    changed += 1

print('Updated:', changed, 'files')
```

```python
# Spot-check samples
def sample_times(files: list[Path], limit: int = 3):
    for path in files[:limit]:
        ts = datetime.fromtimestamp(path.stat().st_mtime, tz=timezone.utc)
        print(path.name, ts.isoformat())

print('Read samples:')
sample_times(read_files)
print('Write samples:')
sample_times(write_files)
```
